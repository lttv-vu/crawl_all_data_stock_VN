```python
import yfinance as yf
from vnstock import *
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import csv
```

```python
data_array = []

# Đọc file CSV và đưa dữ liệu vào mảng
with open('LIST_NEW.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        data_array.append(row)

# In data_array
tickers = [row[0] for row in data_array[1:]]
print(tickers)
```

```python
# cổ đông doanh nghiệp
def crawl_company_large_shareholders(tickers, output_folder):
    
        for symbol in tickers:
            try:
                # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = company_large_shareholders(f'{symbol}')

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}')
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}")
                
# cổ tức
def crawl_dividend_history(tickers, output_folder):
        for symbol in tickers:
            try:
                # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = dividend_history(f'{symbol}')

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}')
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}")
                
                
#New
def crawl_company_news(tickers, output_folder):
    
        for symbol in tickers:
            try:
                 # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = company_news(f'{symbol}', page_size=10, page=0)

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}')
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}")
                
                
#PE_PB
def crawl_stock_evaluation(tickers, output_folder):
    
        for symbol in tickers:
            try:
               # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = stock_evaluation(f'{symbol}', period=1, time_window='D')

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}')
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}")
                
                
                
#Ban lãnh đạo công ty (Mới)                
def crawl_company_officers(tickers, output_folder):
    
        for symbol in tickers:
            try:
                stock_data = company_officers(f'{symbol}', page_size=20, page=0)

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}')
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}")
                
#Báo cáo kinh doanh                    
def crawl_financial_flow(tickers, output_folder):
    
        for symbol in tickers:
            try:
                # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = financial_flow(f'{symbol}', report_type='incomestatement', report_range='quarterly')


                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}') 
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}") 
                  
            
 #Đánh giá               
def crawl_valuation_rating(tickers, output_folder):
    
        for symbol in tickers:
            try:
                # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = valuation_rating(f'{symbol}')

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}') 
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}") 
                  
            
#Tổng quát doanh nghiệp  
def crawl_company_overview(tickers, output_folder):
    
        for symbol in tickers:
            try:
                # Lấy dữ liệu chứng khoán từ yfinance
                stock_data = company_overview(f'{symbol}')

                # Ghi dữ liệu vào DataFrame
                df = pd.DataFrame(stock_data)

                # Lưu DataFrame vào file CSV
                output_file = f'{output_folder}/{symbol}.csv'
                df.to_csv(output_file)

                print(f'Dữ liệu chứng khoán cho mã {symbol} đã được lưu vào file CSV: {output_file}') 
            
            except Exception as e:
                print(f"Lấy dữ liệu lỗi {symbol}: {e}") 
                  
        
```
